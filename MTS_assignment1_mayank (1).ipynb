```python
import yfinance as yf
import pandas as pd
import numpy as np

# Set the indices and equities tickers
indices = ['^GSPC', '^IXIC', '^DJI', '^FTSE', '^N225']  # Example indices
equities = ['AAPL', 'GOOGL', 'AMZN', 'MSFT','INFY']  # Updated list of equities

# Load the historical data for indices
index_data = yf.download(indices, start='2010-01-01', end='2023-05-01', progress=False)['Adj Close']

# Load the historical data for equities
equity_data = yf.download(equities, start='2010-01-01', end='2023-05-01', progress=False)['Adj Close']

# Calculate daily returns for indices
index_returns = index_data.pct_change().dropna()

# Calculate cumulative returns for indices
index_cum_returns = (1 + index_returns).cumprod()

# Calculate max drawdowns for indices
index_max_drawdowns = (index_cum_returns.div(index_cum_returns.cummax()) - 1).min()

# Calculate annualized Sharpe ratios for indexes
risk_free_rate = 0.02  # Example risk-free rate
index_sharpe_ratios = (index_returns.mean() - risk_free_rate) / index_returns.std() * np.sqrt(252)

# Calculate annualized Sortino ratios for equities
index_downside_returns = index_returns[index_returns < 0]
index_sortino_ratios = (index_returns.mean() - risk_free_rate) / index_downside_returns.std() * np.sqrt(252)

# Calculate daily returns for equities
equity_returns = equity_data.pct_change().dropna()

# Calculate cumulative returns for equities
equity_cum_returns = (1 + equity_returns).cumprod()

# Calculate max drawdowns for equities
equity_max_drawdowns = (equity_cum_returns.div(equity_cum_returns.cummax()) - 1).min()

# Calculate annualized Sharpe ratios for equities
risk_free_rate = 0.02  # Example risk-free rate
equity_sharpe_ratios = (equity_returns.mean() - risk_free_rate) / equity_returns.std() * np.sqrt(252)

# Calculate annualized Sortino ratios for equities
equity_downside_returns = equity_returns[equity_returns < 0]
equity_sortino_ratios = (equity_returns.mean() - risk_free_rate) / equity_downside_returns.std() * np.sqrt(252)

# Print the results
print("Index Daily Returns:")
print(index_returns.tail())

print("\nIndex Cumulative Returns:")
print(index_cum_returns.tail())

print("\nIndex Max Drawdowns:")
print(index_max_drawdowns)

print("\nIndex Sharpe Ratios:")
print(index_sharpe_ratios)

print("\nIndex Sortino Ratios:")
print(index_sortino_ratios)

print("\nEquity Daily Returns:")
print(equity_returns.tail())

print("\nEquity Cumulative Returns:")
print(equity_cum_returns.tail())

print("\nEquity Max Drawdowns:")
print(equity_max_drawdowns)

print("\nEquity Sharpe Ratios:")
print(equity_sharpe_ratios)

print("\nEquity Sortino Ratios:")
print(equity_sortino_ratios)
```
